```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

print("Setup complete!")
```

```python
file_path = "C:/Users/48666/Desktop/Expedia_Analysis/data/travel.csv"

#load data
try:
    df=pd.read_csv("C:/Users/48666/Desktop/Expedia_Analysis/data/travel.csv")
    print(" Data loaded successfully!")
    print(f" Dataset shape: {df.shape}")
    print(f" Columns: {list(df.columns)}")
except Exception as e:
    print(f" Error loading data: {e}")
    print(" Check file path and permissions")
```

```python
#Display first 3 rows
df.head(3)

#Basic info
df.info()

#Summary statistics
df.describe().T
```

```python
#Checking missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])

# Handle missing values
if 'Rating' in df.columns:
    df['Rating'].fillna(df['Rating'].median(), inplace=True)
    
print("\n Missing values handled")
```

```python
# Fifth Cell: Visualizations for Available Data
plt.figure(figsize=(18, 20))
plt.suptitle("Expedia Hotel Search Analysis", fontsize=22, y=0.98)

# 1. Top hotel countries
plt.subplot(3, 2, 1)
top_countries = df['hotel_country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Hotel Countries", fontsize=16)
plt.xlabel("Number of Searches", fontsize=12)
plt.ylabel("")
plt.grid(axis='x', linestyle='--', alpha=0.7)

# 2. Booking vs. search
plt.subplot(3, 2, 2)
booking_rate = (df['is_booking'].mean() * 100)
non_booking_rate = 100 - booking_rate
plt.pie([booking_rate, non_booking_rate], 
        labels=['Bookings', 'Searches Only'], 
        autopct='%1.1f%%', 
        colors=['#ff9999','#66b3ff'],
        explode=(0.1, 0))
plt.title(f"Booking Rate: {booking_rate:.1f}%", fontsize=16)

# 3. Mobile vs. desktop
plt.subplot(3, 2, 3)
mobile_rate = (df['is_mobile'].mean() * 100)
device_data = [mobile_rate, 100 - mobile_rate]
sns.barplot(x=['Mobile', 'Desktop'], y=device_data, palette="rocket")
plt.title("Device Usage", fontsize=16)
plt.ylabel("Percentage of Searches", fontsize=12)
for i, v in enumerate(device_data):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center', fontsize=12)

# 4. Travel group composition
plt.subplot(3, 2, 4)
adults_mean = df['srch_adults_cnt'].mean()
children_mean = df['srch_children_cnt'].mean()
group_data = [adults_mean, children_mean]
sns.barplot(x=['Adults', 'Children'], y=group_data, palette="mako")
plt.title("Average Group Composition", fontsize=16)
plt.ylabel("Average Count", fontsize=12)
for i, v in enumerate(group_data):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', fontsize=12)

# 5. Package deals
plt.subplot(3, 2, 5)
package_rate = (df['is_package'].mean() * 100)
package_data = [package_rate, 100 - package_rate]
plt.pie(package_data, 
        labels=['Package Deals', 'Non-Package'], 
        autopct='%1.1f%%', 
        colors=['#ffcc99','#99ff99'],
        shadow=True)
plt.title("Package Deals", fontsize=16)

# 6. Search channels
plt.subplot(3, 2, 6)
channel_counts = df['channel'].value_counts().head(10)
sns.barplot(x=channel_counts.values, y=channel_counts.index.astype(str), palette="flare")
plt.title("Top Search Channels", fontsize=16)
plt.xlabel("Number of Searches", fontsize=12)
plt.ylabel("Channel ID", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('../visualizations/hotel_search_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Time-based analysis
df['search_date'] = pd.to_datetime(df['date_time']).dt.date
daily_searches = df.groupby('search_date').size()

plt.figure(figsize=(12, 6))
daily_searches.plot(kind='line', color='purple')
plt.title("Daily Search Volume", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Searches", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../visualizations/daily_searches.png', dpi=300)
plt.show()

# Booking rate by traveler type
plt.figure(figsize=(10, 6))
booking_by_type = df.groupby('traveler_type')['is_booking'].mean().sort_values(ascending=False) * 100
sns.barplot(x=booking_by_type.values, y=booking_by_type.index, palette="viridis")
plt.title("Booking Rate by Traveler Type", fontsize=16)
plt.xlabel("Booking Rate (%)", fontsize=12)
plt.ylabel("")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../visualizations/booking_by_type.png', dpi=300)
plt.show()
```

```python
# Cell 6: Key Metrics Calculation
# Calculate key business metrics
total_searches = len(df)
booking_rate = df['is_booking'].mean() * 100
mobile_share = df['is_mobile'].mean() * 100
package_share = df['is_package'].mean() * 100
avg_adults = df['srch_adults_cnt'].mean()
avg_children = df['srch_children_cnt'].mean()
top_country = df['hotel_country'].value_counts().idxmax()

# Traveler type analysis
def classify_traveler(row):
    if row['srch_adults_cnt'] == 1 and row['srch_children_cnt'] == 0:
        return 'Solo'
    elif row['srch_adults_cnt'] == 2 and row['srch_children_cnt'] == 0:
        return 'Couple'
    elif row['srch_children_cnt'] > 0:
        return 'Family'
    else:
        return 'Group'

df['traveler_type'] = df.apply(classify_traveler, axis=1)
traveler_distribution = df['traveler_type'].value_counts(normalize=True) * 100

print(f"""
 Key Business Insights:
1. Total searches: {total_searches:,}
2. Booking conversion rate: {booking_rate:.2f}%
3. Mobile search share: {mobile_share:.2f}%
4. Package deal share: {package_share:.2f}%
5. Most popular hotel country: {top_country}
6. Average group size: {avg_adults:.2f} adults, {avg_children:.2f} children

Traveler Type Distribution:
- Solo: {traveler_distribution.get('Solo', 0):.2f}%
- Couple: {traveler_distribution.get('Couple', 0):.2f}%
- Family: {traveler_distribution.get('Family', 0):.2f}%
- Group: {traveler_distribution.get('Group', 0):.2f}%
""")
```
